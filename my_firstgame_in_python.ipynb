```python
from IPython.display import clear_output

boat_side='Right'
missionaries_on_right=3
cannibals_on_right=3
missionaries_on_left=0
cannibals_on_left=0
boat=' |---B| '

def display():

    user_interface = '' 

    for i in range(missionaries_on_left):
      user_interface=user_interface+ '\U0001f482' 
    
    for i in range(cannibals_on_left):
      user_interface=user_interface+  '\U0001f479' 
    
    if boat_side!='Right':   
     user_interface=user_interface+'\U0001f6A2'

    for i in range(5):
      user_interface=user_interface+  '\U0001f30a'
    
    if boat_side=='Right':
      user_interface=user_interface+'\U0001f6A2'
    
    for i in range(missionaries_on_right):
      user_interface=user_interface+ '\U0001f482' 

    for i in range(cannibals_on_right):
      user_interface=user_interface+  '\U0001f479' 

    print(user_interface)
    
while True:
    
    print('M='+str(missionaries_on_left)+' C=' +str(cannibals_on_left)+str(boat)+
      ' M='+str(missionaries_on_right)+' C=' +str(cannibals_on_right), end='\n' ) 
    
    display()
    
    print('enter no of missionaries upto 2')
    missionaries=int(input())
    
    print('enter no of cannibals upto '+ str(2-missionaries))
    cannibals=int(input())
    
    x=missionaries+cannibals
    if x > 2 :
      print('invalid move')
      continue
      
    
    if boat_side=='Right':
        
        if missionaries_on_right <missionaries or cannibals_on_right < cannibals:
          print('invalid move')
          continue
    
        missionaries_on_right=missionaries_on_right-missionaries
        cannibals_on_right=cannibals_on_right-cannibals
        
        missionaries_on_left=missionaries_on_left+missionaries
        cannibals_on_left=cannibals_on_left+cannibals
        boat_side='Left'
        boat=' |B------------| '
        clear_output()
        
    else:
       
        if missionaries_on_left < missionaries or cannibals_on_left< cannibals:
          print('invalid move')
          continue
          
        missionaries_on_left= missionaries_on_left-missionaries
        cannibals_on_left=cannibals_on_left-cannibals
        
        missionaries_on_right=missionaries_on_right+missionaries
        cannibals_on_right=cannibals_on_right+cannibals
        boat_side='Right'
        boat=' |------------B| '
        
        clear_output()
                
    if missionaries_on_right < cannibals_on_right and missionaries_on_right > 0 or missionaries_on_left < cannibals_on_left and missionaries_on_left > 0 :
         print('YOU LOSE')
         print('GAME OVER')
         break
    
    if missionaries_on_left == 3 and cannibals_on_left==3 :
         print('YOU WIN') 
         print('GAME OVER')
         break
       
```
